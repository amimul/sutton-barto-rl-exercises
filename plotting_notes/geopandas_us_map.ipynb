```python
from eda_imports import *
import geopandas
```

```python
# data downloaded from http://eric.clst.org/tech/usgeojson/
files = glob.glob('./geojson_data/*')
files = sorted([_ for _ in files if os.path.islink(_)])
```

```python
files.remove('./geojson_data/us_counties.json')
```

```python
files
```

```python
fig, axes = plt.subplots(4, 1, figsize=(20, 50), sharex=True, sharey=True)
axes = axes.ravel()

for k, f in tqdm(enumerate(files)):
    df = geopandas.read_file(f)
    
    ax = axes[k]
#     df.plot(figsize=(16, 9), legend=True, color='white', edgecolor='black', ax=ax)
    df.plot(figsize=(16, 9), ax=ax)
    ax.set_xlim(-200, -50)
    ax.set_title(f)
    ax.grid()
plt.tight_layout()
```
