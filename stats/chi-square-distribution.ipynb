```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Verify that $\chi^2$ distributed random variable ($Q$) is just a sum of squares of a number ($n$, or degree of freedom) of normally distributed random variables ($X_i$)

\begin{align*}
Q &= \sum_{i=1}^{n}X_i^2 \\
Q &\sim \chi^2(n)
\end{align*}

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
axes = axes.ravel()

bins = np.linspace(0, 20, 100)
for k, dof in enumerate(range(1, 7)):
    
    # number of data points to simulate
    num = int(1e6)
    
    # Q
    simulated_from_norm_dist = sum(np.random.standard_normal(num) ** 2 for _ in range(dof))
    
    # reference normal distribution, for comparison purpose
    ref = np.random.chisquare(df=dof, size=num)

    ax = axes[k]
    _ = ax.hist(simulated_from_norm_dist, bins=bins, histtype='step', label='dof={0}'.format(dof))
    _ = ax.hist(ref, bins=bins, histtype='step', label='ref')
    ax.legend()
    ax.set_title('{0} degree-of-freedom $\chi^2$ distribution'.format(dof))
plt.tight_layout()
```

# Put $\chi^2$ distributions of different degrees of freedom together

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

bins = np.linspace(0, 20, 100)
for k, dof in enumerate(range(1, 7)):
    # number of data points to simulate
    num = int(1e6)
    ref = np.random.chisquare(df=dof, size=num)

    _ = ax.hist(ref, bins=bins, histtype='step', label='dof={0}'.format(dof))
ax.legend()
ax.set_ylim(0, ax.get_ylim()[1] / 2)
```
