```python
def exponential_loss(yF):
    return np.exp(-yF)
```

```python
def binomial_deviance(yF):
    # add constant - np.log(2) + 1 so that it passes the (0, 1) point
    return np.log(1 + np.exp(- 2 * yF)) - np.log(2) + 1
```

```python
def squared_error_loss(delta_y_F):
    return delta_y_F ** 2
```

```python
def misclassification(yF):
    return (yF < 0).astype(int)
```

This plot is consistent with Figure 10.4 at Page 347 from https://web.stanford.edu/~hastie/Papers/ESLII.pdf

```python
fig = plt.figure(figsize=(8, 6))

yFs = np.arange(-2, 2, 0.01)
plt.plot(yFs, exponential_loss(yFs), label='Exponential', color='blue')
plt.plot(yFs, binomial_deviance(yFs), label='Binomial deviance', color='orange')
plt.plot(yFs, misclassification(yFs), label='Misclassification', color='gray')

Fs = np.arange(0, 1, 0.01)
ys = np.array([-1, 1])
delta_y_F = np.concatenate([ys[0] - Fs, ys[1] - Fs])
_yFs = np.concatenate([ys[0] * Fs, ys[1] * Fs])
idxes = np.argsort(_yFs)
delta_y_F = delta_y_F[idxes]

plt.plot(np.sort(_yFs), squared_error_loss(delta_y_F), label='Dquared error', color='red')

plt.legend()
plt.ylim(-0.5, 4)
plt.grid()
plt.xlim(-2, 2)
plt.xlabel('y * F')
plt.ylabel('Loss')
```
