# Make up an artificial loss surface

```python
xs, ys = np.mgrid[-3:3:.1, -3:3:.1]

pos = np.empty(xs.shape + (2,))

pos[:, :, 0] = xs
pos[:, :, 1] = ys

zs = mixed_gaussian(pos)
# plt.contourf(xs, ys, zs)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.gca(projection='3d')
ax.plot_surface(xs, ys, zs, rstride=1, cstride=1, cmap=mpl.cm.coolwarm,
                linewidth=0, antialiased=False)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
```
